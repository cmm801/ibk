```python
def get_datetimes(ts):
    """Convert the numpy.datetime64 into Python datetime objects."""
    return get_datetimes_from_numpy(ts.index)

def get_datetimes_from_numpy(np_datetimes):
    return [x.to_pydatetime() for x in pd.DatetimeIndex(np_datetimes)]

def _is_repeated(ts, n_times=3):
    """Find the indices of data that has been repeated at least n_times.
       This function is used to clean data from IB, which seems to include
          periods of repeated stale data.
    """
    is_repeated = np.ones_like(ts.shape[0], dtype=np.bool)
    for s in range(n_times):
        is_repeated = is_repeated & (ts.close == ts.close.shift(s+1))
    return is_repeated

def find_profitable_periods(ts, lower_limit, upper_limit, window_size=None):
    
    # Check bars are all the same size, if we are looking at a fixed window size
    datetimes = get_datetimes(ts)
    bar_sizes = [(datetimes[t] - datetimes[t-1]).seconds for t in range(1, len(datetimes))]    
    if window_size is not None:
        assert len(set(bar_sizes)) == 1, "The bar sizes must all be the same."
    bar_size = min(bar_sizes)
    
    y = np.nan * np.ones_like(ts.average, dtype=np.float32)
    T = ts.shape[0]
    for t_0 in range(T):
        # Get the starting price, and the upper/lower target prices
        start_price = ts.close.values[t_0]
        upper_price = start_price * (1 + upper_limit)
        lower_price = start_price * (1 + lower_limit)

        # Determine how many periods into the future we will check
        if window_size is None:
            t_max = T
        else:
            t_max = min(T, t_0 + window_size)
            
        t_f = t_0 + 1
        if t_f >= T:
            continue            
        y[t_f] = 0
        finished = False
        while t_f < t_max and not finished:
            is_above_target = ts.high.values[t_f] >= upper_price
            is_below_target = ts.low.values[t_f] <= lower_price
            if is_above_target and is_below_target:
                # Both upper and lower prices are breached in this bar.
                # Higher frequency data is needed to distinguish which
                #    occured first.
                y[t_0] = 0
                finished = True
            elif is_above_target:
                y[t_0] = (datetimes[t_f] - datetimes[t_0]).seconds
                finished = True
            elif is_below_target:
                y[t_0] = -1 * (datetimes[t_f] - datetimes[t_0]).seconds
                finished = True
            else:
                t_f += 1

    return y

def get_contiguous_periods(input_ts, min_period_length=120, max_repeat=75):
    """Split the time series into a list of contiguous periods.
    """
    # Remove stale data with repeated prices
    repeated_count = get_repeated_counts(input_ts)
    ts = input_ts[repeated_count <= max_repeat]
    
    # Find the locations where the time series is not contiguous
    datetimes = get_datetimes(ts)
    dt = np.array([(datetimes[t] - datetimes[t-1]).seconds for t in range(1, len(datetimes))])
    bar_size = dt[0]
    idx_split = 1 + np.where(dt != bar_size)[0]

    ts_periods = []
    idx_last = 0
    for idx in idx_split:
        if idx - idx_last >= min_period_length:
            ts_periods.append(ts.iloc[idx_last:idx,:])
        idx_last = idx

    # Add the remaining contiguous time series to the list
    ts_periods.append(ts.iloc[idx_last:,:])
    return ts_periods

def get_repeated_counts(ts):
    """Count how many times a stale price is repeated 
         within each group in the time series.
       """
    T = ts.shape[0]
    repeated_count = np.zeros((T,))
    
    # Increment the count each time a new stale price is added
    for t in range(1, T):
        if ts.close.values[t] == ts.close.values[t-1]:
            repeated_count[t] = repeated_count[t-1] + 1

    # Set each count in the same stale group to the same value
    for t in reversed(range(T-1)):
        if repeated_count[t] > 0 and repeated_count[t+1] > 0:
            repeated_count[t] = repeated_count[t+1]
    
    return repeated_count

def get_volume_weighted_time_series(ts):
    """Create a time series indexed by the volume instead of time
    """
    ts_vol = ts.copy()
    ts_vol = ts_vol.reset_index()
    ts_vol = ts_vol[ts_vol.volume > 0]
    V = ts_vol.volume.cumsum()
    ts_vol.index = pd.Index(V, name='volume')
    return ts_vol

def calc_ratio_up_down_moves(price, window):
    price_ind = (price > price.shift(1)).astype(np.float32)
    return price_ind.rolling(window=window).sum() / window

def calc_ratio_up_down_moves_by_volume(price, volume, window):
    cum_vol = volume.cumsum()
    tot_vol = cum_vol - cum_vol.shift(window_size)
    price_ind = (price > price.shift(1)).astype(np.float32)
    vol_wtd_ind = price_ind * volume
    return vol_wtd_ind.rolling(window=window).sum() / tot_vol

def calc_ema(input_ts, alpha, weights=None):
    """Comuputes the Exponential Moving Average (EMA).
       Takes numpy array as input.
    """    
    if weights is None:
        weights = np.ones_like(input_ts, dtype=np.float32)
        
    ema = np.zeros_like(input_ts, dtype=np.float32)
    denom = np.zeros_like(input_ts, dtype=np.float32)
    ema[0] = input_ts[0]
    denom[0] = weights[0]
    for t in range(1, input_ts.shape[0]):
        denom[t] = weights[t] + (1 - alpha) * denom[t-1]
        alpha_wtd = weights[t] / denom[t]
        ema[t] = (1-alpha_wtd) * ema[t-1] + alpha_wtd * input_ts[t]
    return ema

def calc_smma(input_ts, window, weights=None):
    """Comuputes the Smoothed modified moving average.
       Takes numpy array as input.    
    """
    return calc_ema(input_ts, alpha=1/window, weights=weights)

def calc_rsi(price, window, weights=None):
    up = np.zeros_like(price, dtype=np.float32)
    down = np.zeros_like(price, dtype=np.float32)
    dp = price - price.shift(1)
    up[dp > 0] = dp[dp > 0]
    down[dp < 0] = -dp[dp < 0]

    up_smma = calc_smma(up, window=window, weights=weights)
    down_smma = calc_smma(down, window=window, weights=weights)
    RS = up_smma / (down_smma + 1e-10)
    rsi = 100 * (1 - 1/(1+RS))
    return rsi

def calc_macd(price, window_fast, window_slow, weights=None):
    assert window_fast < window_slow, 'Fast window must be smaller than slow window.'
    
    prc_slow = calc_smma(price, window=window_slow, weights=weights)
    prc_fast = calc_smma(price, window=window_fast, weights=weights)
    return prc_fast - prc_slow
```

```python
def calc_crossover_positions(window_slow, window_fast, weights=None, threshold=0.0025 ):
    smma_fast = calc_smma(input_ts, window=window_fast, weights=weights)
    smma_slow = calc_smma(input_ts, window=window_slow, weights=weights)
    
    T = input_ts.shape[0]
    positions = np.zeros((T,), dtype=np.float32)
    for t in range(window_slow-1, T):
        if smma_fast[t] > smma_slow[t] * (1 + threshold):
            positions[t] = +1
        elif smma_fast[t] < smma_slow[t] * (1 - threshold):
            positions[t] = -1
        else:
            positions[t] = positions[t-1]
    return positions
    
def run_strategy(input_ts, positions):    
    T = input_ts.shape[0]
    mv = np.zeros_like(input_ts, dtype=np.float32)
    mv[:window_slow] = 1.0
    for t in range(window_slow-1, T):
        rtn = -1 + input_ts[t] / input_ts[t-1]
        if input_ts[t-1] == 0 or np.isnan(input_ts[t-1]):
            print('here')
        mv[t] = mv[t-1] * (1 + positions[t-1] * rtn)

    rtns = mv[1:] / mv[:-1] - 1
    perf_stats = dict(tot_rtn=mv[-1] - 1, std_dev=rtns.std(), mdd=min(mv) - 1)
    return mv, perf_stats
```

```python
window_fast = 1
window_slow = 50
threshold = 0.0025

T_train = int(0.75 * ts_vol.shape[0])
input_ts = ts_vol.average.head(T_train).values
#weights = ts_vol.volume.values
weights = None

positions = calc_crossover_positions(window_slow=window_slow, window_fast=window_fast, 
                                     weights=weights, threshold=threshold)
mv, perf_stats = run_strategy(input_ts, positions)
print(perf_stats)
```

```python
fast_windows = [1, 3, 5, 10, 15, 20, 30]
slow_windows = [30, 50, 60, 90, 120, 240]
thresholds = [0, 0.001, 0.0025, 0.005, 0.0075, 0.01]
weights = None

stats = []
t0 = time.time()
for threshold in thresholds:
    for fast in fast_windows:
        for slow in slow_windows:
            positions = calc_crossover_positions(window_slow=slow, window_fast=fast, 
                                                 weights=weights, threshold=threshold)
            mv, perf = run_strategy(input_ts, positions)
            perf['slow'] = slow
            perf['fast'] = fast
            perf['threshold'] = threshold
            perf['n_trades'] = np.abs(positions[1:] - positions[:-1]).sum()
            stats.append(perf)
print(time.time() - t0)
```

```python

class RequestIdManager(object):
    """Class to help manage the request ids.
    """
    def __init__(self):
        # Keep track of request IDs and human-readable names/ids
        super().__init__()
        print('request_id')
        self.req_id = None
        self._req_ids_to_string_ids = dict()
        self._string_ids_to_req_ids = dict()
    
    def _get_next_req_id(self):
        """Retrieve the current class variable req_id and increment
        it by one.

        Returns (int) current req_id
        """
        current_req_id = self.req_id
        self.req_id += 1
        return current_req_id

    def _log_request_id(self, req_id, label):
        """Keep track of request IDs and their human-readable labels."""
        self._req_ids_to_string_ids[req_id] = label
        self._string_ids_to_req_ids[label] = req_id

    def _get_req_id_from_label(self, label):
        return self._string_ids_to_req_ids[label]
    
    def _get_label_from_req_id(self, req_id):
        return self._req_ids_to_string_ids[req_id]
    
    def _is_new_request(self, label):
        """Check if a label has been logged with a request."""
        return label not in self._req_ids_to_string_ids
    

class A1(object):
    def __init__(self):
        super().__init__()        
        self.a1 = 1
        
    def get_a1(self):
        print(self.a1)
        
class A2(object):
    def __init__(self):
        super().__init__()        
        self.a2 = 2
    
    def get_a2(self):
        print(self.a2)
        
        
class B(A1, A2):
    __metaclass__ = RequestIdManager
    
    def __init__(self):
        A1().__init__()
        
class C(A1, A2):
    __metaclass__ = RequestIdManager
    
    def __init__(self):
        super().__init__()
```
